```python
def minutes_to_hours(minute):
    return str(timedelta(minutes=minute))[:-3]

res_applaunch_time=list(map(minutes_to_hours,app_launch_df['Minute_Of_Day']))
app_launch_df['Time']=res_applaunch_time
applaunch_sec=["%02d"% i for i in app_launch_df['Second']]
before_colon_applaunch_time=[a.split(':')[0] for a in app_launch_df['Time']]
before_colon_2digit_applaunch_time=["%02d"% i for i in pd.to_numeric(before_colon_applaunch_time)]
after_colon_applaunch_time=[b.split(':')[1] for b in app_launch_df['Time']]
original_applaunch_time=[a+b for a,b in zip(before_colon_2digit_applaunch_time,after_colon_applaunch_time)]
time_with_applaunch_sec=[a+str(b) for a,b in zip(original_applaunch_time,applaunch_sec)]
Date_applaunch=[str(a)+b for a,b in zip(app_launch_df['Date'],time_with_applaunch_sec)]
date_applaunch=[]
for i in Date_applaunch:
    date_applaunch.append(datetime(year=int(i[0:4]), month=int(i[4:6]), day=int(i[6:8]), hour=int(i[8:10]), minute=int(i[10:12]), second=int(i[12:14])))

app_launch_df['LaunchDate']=date_applaunch
```

```python
app_launch_df['PrevLaunch'] = app_launch_df.groupby('UserId')['LaunchDate'].shift(1)
```

```python
app_launch_df['timediff'] = (app_launch_df['LaunchDate']- app_launch_df['PrevLaunch']).dt.seconds
mas_df = mas_df.merge(app_launch_df[['timediff','UserId']].groupby('UserId').mean().fillna(0),
                              how='left',
                              left_on = 'UserId',
                              right_index=True
                             )
mas_df=mas_df.fillna(0)
```

```python
appuninstalled=app_uninstall_df.drop_duplicates()
applaunched=app_launch_df.drop_duplicates()
```

```python
 df_reinstalled= appuninstalled.groupby('UserId')['Date'].max().to_frame().merge(applaunched.groupby('UserId')['Date'].max().to_frame(),
                                                                left_index=True,
                                                                right_index=True,how='outer',
                                                                suffixes=['_uninstalled','_launched'] )
    
df_reinstalled['Reinstalled'] = df_reinstalled['Date_uninstalled'] < df_reinstalled['Date_launched']
df_reinstalled['Reinstalled'] = df_reinstalled['Reinstalled'].astype(int)
    
mas_df = mas_df.merge( df_reinstalled[['Reinstalled']] ,
                              how='left',
                              left_on = 'UserId',
                              right_index =True
                             )
```

```python
videostarted=video_started_df.drop_duplicates()
videodetails=video_details_df.drop_duplicates()
```

```python
mas_df=mas_df.merge(video_started_df.groupby(['UserId','Genre'])['VideoId'].nunique().unstack().fillna(0),
                            how='left',
                            left_on = 'UserId',
                            right_index =True,
                            suffixes = ['','_video_ratio']
                             )
video_per_genre = videodetails.groupby(['Genre'])['VideoId'].nunique()
    
for col in mas_df.columns:
    try:
        if col.endswith('_video_ratio'):
            genre_total_vid = video_per_genre.loc[col.replace('_video_ratio','')]
            mas_df[col] = mas_df[col] / genre_total_vid
    except:
        print("An exception occurred")
                  
```
